```python
# Function to issue books
def issue_books(student_name, books):
  print(f"\n {student_name} borrowed:")
  for book in books:
    print("-", book)
# Function to calculate fine
def calculate_fine(days_late):
  fine_per_day = 5
  return days_late * fine_per_day
# Function to generate report
def student_report(student_name, books, days_late):
  issue_books(student_name, books)
  fine = calculate_fine(days_late)
  print(f"Fine to be paid: ${fine}\n")
# Example usage
student_report("Rahul", {"Python Basics", "Data Science 101"}, 3)
student_report("Priya", ['Machine Learning'], 0)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#Example data
data = [10,12,13,12,11,200,14,13,15,12]
sns.boxplot(data)
plt.show()
#detecting using IQR
Q1, Q3 = np.percentile(data, [25, 75])
IQR = Q3 - Q1
lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
outliers = [x for x in data if x < lower or x > upper]
print("Outliers detected:", outliers)
```
